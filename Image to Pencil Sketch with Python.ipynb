```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/dog.jpg')
cv2_imshow(img)
```

```python
img1= cv2.imread('/content/dog.jpg') 
blur_image = cv2.blur(img1,(100,100))
print("diminished dimensions",blur_image.shape) 
cv2_imshow(blur_image)
```

```python
from google.colab.patches import cv2_imshow
#** _imshow got updated 
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)

cv2.waitKey(0)
```

```python
from google.colab.patches import cv2_imshow

inverted_image = 255 - gray_image
print('Inverted Image ')
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
print("Sketch")
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

```python
print("\n original image \n")
cv2_imshow(img)
print("\n Pencil Image\n")
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

**The pencil sketch of a is ready!**
